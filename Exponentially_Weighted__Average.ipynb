```python
def ewa(x):
  vti=0
  xnew=[]
  for i in range(len(x)):
    vti=0.8*vti+0.2*x[i]
    # vti=vti/(1-(0.9**(i+1)))
    # print(vti)
    xnew.append(vti)
  return xnew
```

```python
plt.plot(range(len(df[0])),ewa(df[3]))
# ewa(df[3]).unique
```
